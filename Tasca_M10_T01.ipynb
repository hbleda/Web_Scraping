```python
# Cargamos la página web
url_citas = 'http://quotes.toscrape.com'
page_quote = requests.get(url_citas)
print(page_quote.text)
```

<div class="alert alert-info alert-info">   
Creamos una función que itere sobre las distintas páginas con citas para obtenerlas todas
</div>

```python
def web_scraping_soup(url):
    frases = []
    autores = []
    etiquetas = []

    while True:
        # Enviamos una petició GET a la URL
        response = requests.get(url)

        # Creamos un objecto BeautifulSoup
        soup_citas = BeautifulSoup(response.content, "html.parser")

        # Encontramos los elementos div
        citas = soup_citas.find_all("div", class_="quote")

        # Iteramos sobre cada cita
        for cita in citas:
            # Sacamos el texto de la cita
            texto = cita.find("span", class_="text").text

            # Sacamos el autor de la cita
            autor = cita.find("small", class_="author").text

            # Sacamos los tags asociados a la cita
            etiqueta = cita.find_all("a", class_="tag")
            etiquetas_cita = [tag.text for tag in etiqueta]

            # Añadimos cita, autor y tags a la lista creadas anteriormente
            frases.append(texto)
            autores.append(autor)
            etiquetas.append(etiquetas_cita)

        # Comprovamos si hay una página más
        next_page = soup_citas.find("li", class_="next")
        if next_page is None:
            # Si no hay más paginas salimos del bucle
            break
        else:
            # Si hay más paginas, recnstruimos la url
            next_url = urljoin(url, next_page.find("a")["href"])
            url = next_url

    # Creamos un diccionario
    data = {"Frases": frases, "Autors": autores, "Etiquetes": etiquetas}

    # Creamos un DataFrame a partir del diccionario
    df = pd.DataFrame(data)

    # Retornem el DataFrame
    return df
```

```python
# Llamamos a la función creada para "raspar" la web
df_citas_soup = web_scraping_soup(url_citas)
df_citas_soup
```

```python
# Guardamos los datos en un csv
df_citas_soup.to_csv("citas_soup.csv", index=False)
```

<div class="alert alert-info alert-info">   
Repetimos el web scraping de la web http://quotes.toscrape.com  con Selenium.
</div>

```python
#Creamos un función para realizar el scraping
def web_scraping_sele(url):
    frases = []
    autores = []
    etiquetas = []

    while True:
        # Creamos una instancia del navegador Chrome
        options = Options()
        options.add_argument('--headless')
        options.add_argument('--incognito')
        driver = webdriver.Chrome(options=options)
        # Abrimos la url en el navegador
        driver.get(url)  
        # Esperamos a que se cargue la página completamente
        driver.implicitly_wait(10)

        # Guardamos todos los elementos de las citas
        citas = driver.find_elements(By.CSS_SELECTOR, '.quote')

        # Iteramos sobre cada cita
        for cita in citas:
            # Sacamos el texto de la cita
            texto = cita.find_element(By.CSS_SELECTOR, '.text').text

            # Sacamos el autor de la cita
            autor = cita.find_element(By.CSS_SELECTOR, '.author').text

            # Sacamos los tags asociados a la cita
            etiquetas_citas = [tag.text for tag in cita.find_elements(By.CSS_SELECTOR, '.tag')]

            # Agregamos la cita, el autor y las etiquetas a las listas correspondientes
            frases.append(texto)
            autores.append(autor)
            etiquetas.append(etiquetas_citas)

        try:
            # Comprobamos si hay una página más y si es así, modificamos la url
            next_button = driver.find_element(By.CSS_SELECTOR, 'li.next a')
            url = next_button.get_attribute('href')
        except NoSuchElementException:
            # Si no hay más páginas, salimos del bucle
            break

        driver.quit()  # Close the browser

    # Creamos un diccionario
    data = {"Frases": frases, "Autores": autores, "Etiquetas": etiquetas}

    # Creamos un DataFrame a partir del diccionario
    df = pd.DataFrame(data)

    # Retornamos el DataFrame
    return df
```

```python
# Llamamos a la función creada para "raspar" la web
df_citas_sele = web_scraping_sele(url_citas)
df_citas_sele
```

```python
# Guardamos los datos en un csv
df_citas_sele.to_csv("citas_sele.csv", index=False)
```

```python
# Cerramos la ventana del navegador
driver.quit()
```

```python
import docx

def read_word_document(file_path):
    try:
        # Abrir el documento de Word
        doc = docx.Document(file_path)

        # Leer el contenido del documento
        content = []
        for paragraph in doc.paragraphs:
            content.append(paragraph.text)

        # Unir las líneas para obtener el texto completo
        text = "\n".join(content)
        
        return text
    except Exception as e:
        print(f"Error al leer el documento: {str(e)}")
        return None

if __name__ == "__main__":
    # Ruta del documento de Word que deseas leer
    file_path = "C:/Users/USER/Documents/Data Science/Sprint 10. Web scraping i automatització/Tasca M10 T01/descripcion_dataset.docx"
    content = read_word_document(file_path)
    if content:
        print(content)
```

```python
# Creamos una instancia del navegador Chrome
options = Options()
options.add_argument('--headless')
options.add_argument('--incognito')
driver = webdriver.Chrome(options=options)
```

```python
url = "https://es.wikipedia.org/wiki/Anexo:Pel%C3%ADculas_con_las_mayores_recaudaciones"
driver.get(url)
```

```python
tabla = driver.find_element(By.CSS_SELECTOR, ".wikitable")
print(tabla.text)
```

```python
peliculas = []
distribuidoras = []
taquillas_EEUU = []
taquillas_fuera_EEUU = []
recaudaciones_mundial = []
presupuestos = []
anyos_estreno = []
directores = []
```

```python
files = tabla.find_elements(By.TAG_NAME, "tr")[1:]
```

```python
for fila in files:

    celes = fila.find_elements(By.TAG_NAME, "td")
    pelicula = celes[0].text.strip()
    distribuidora = celes[1].text.strip()
    taquilla_EEUU = celes[2].text.strip()
    taquilla_fuera_EEUU = celes[3].text.strip()
    recaudacion_mundial = celes[4].text.strip()
    presupuesto = celes[5].text.strip()
    anyo_estreno = celes[6].text.strip()
    director = celes[7].text.strip()
    
    peliculas.append(pelicula)
    distribuidoras.append(distribuidora)
    taquillas_EEUU.append(taquilla_EEUU)
    taquillas_fuera_EEUU.append(taquilla_fuera_EEUU)
    recaudaciones_mundial.append(recaudacion_mundial)
    presupuestos.append(presupuesto)
    anyos_estreno.append(anyo_estreno)
    directores.append(director)
```

```python
driver.quit()
```

```python
data = {"Película": peliculas, "Distribuidora": distribuidoras, "Taquilla en EEUU": taquillas_EEUU,
        "Taquilla fuera EEUU": taquillas_fuera_EEUU, "Recuaudación Mundial": recaudaciones_mundial,
        "Presupuesto": presupuestos,"Año Estreno": anyos_estreno,"Director": directores}

mas_taquilleras_df = pd.DataFrame(data)
mas_taquilleras_df
```

```python
# Guardamos los datos en un csv
mas_taquilleras_df.to_csv("mas_taquilleras.csv", index=False)
```
